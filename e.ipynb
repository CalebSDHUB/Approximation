```python
import numpy as np
import matplotlib.pyplot as plt
```

### Euler approximation function

```python
def e(x, N):

  array = []

  for n in range(1,N):
    array.append(float(pow(x,n)) / float(np.math.factorial(n)))

  return 1 + sum(array)
```

### Euler approximation of the natural constant 'e' (Visualization)

```python
errors = []
results = []

x = 1 # Your chosen power number of e
N = 20 # Number of iterations

for n in range(1,N):

  value = e(x, n)

  results.append(value)
  errors.append(abs(np.e - value))

plt.plot(np.linspace(x, N, N-1),results, 'b', label = "My approx.")
plt.plot(np.linspace(x, N, N-1),np.full((N-1,), np.e), 'g', label = "The constant e.")
plt.xlabel('X - Iterations')
plt.ylabel('Y - Values')
plt.title('Natural constant (e) approximation')
plt.legend()
fig = plt.gcf()
fig.set_dpi(100)
```

### The error approximation of e

```python
plt.plot(np.linspace(x, N, N-1), errors, 'r', label = "Error")
plt.xlabel('X - Iterations')
plt.ylabel('Y - Residual Values')
plt.title('Natural constant (e) approximation residuals')
plt.legend()
fig = plt.gcf()
fig.set_dpi(100)

print(f"Residuals: {errors}")
```

#### Conclusion: After 6 iterations we have a noticable optimum (Residual: 0.0016151617923783057), but for further fine optimization we see that after 18 iterations the values does not really change (Residual: 4.440892098500626e-16)
